```python
# Programa: Soporte Técnico Interactivo
# Nombre: Arevalo Rafael
# NL = 3

# Clase
class SoporteTecnico:
    def __init__(self, nombre, telefono, edad):
        self.nombre = nombre
        self.telefono = telefono
        self.edad = edad

    # Funciones
    def reparar_equipo(self):
        print(self.nombre, "esta reparando un equipo de computo")

    def instalar_software(self):
        print(self.nombre, "esta instalando software")

    def diagnosticar(self):
        print(self.nombre, "esta realizando un diagnostico")


# Crear objetos
tecnico1 = SoporteTecnico("Carlos Perez", "6561112233", 30)
tecnico2 = SoporteTecnico("Luis Ramirez", "6564445566", 25)


# Menú
opcion = 0

while opcion != 6:
    print("\n--- MENU SOPORTE TECNICO ---")
    print("1. Ver datos Tecnico 1")
    print("2. Ver datos Tecnico 2")
    print("3. Reparar equipo")
    print("4. Instalar software")
    print("5. Diagnosticar")
    print("6. Salir")

    opcion = int(input("Elige una opcion: "))

    if opcion == 1:
        print("\nNombre:", tecnico1.nombre)
        print("Telefono:", tecnico1.telefono)
        print("Edad:", tecnico1.edad)

    elif opcion == 2:
        print("\nNombre:", tecnico2.nombre)
        print("Telefono:", tecnico2.telefono)
        print("Edad:", tecnico2.edad)

    elif opcion == 3:
        t = int(input("Elige tecnico (1 o 2): "))
        if t == 1:
            tecnico1.reparar_equipo()
        else:
            tecnico2.reparar_equipo()

    elif opcion == 4:
        t = int(input("Elige tecnico (1 o 2): "))
        if t == 1:
            tecnico1.instalar_software()
        else:
            tecnico2.instalar_software()

    elif opcion == 5:
        t = int(input("Elige tecnico (1 o 2): "))
        if t == 1:
            tecnico1.diagnosticar()
        else:
            tecnico2.diagnosticar()

    elif opcion == 6:
        print("Programa terminado")
        print("Programa generado por Arevalo Rafael NL = 3")

    else:
        print("Opcion incorrecta")
```
